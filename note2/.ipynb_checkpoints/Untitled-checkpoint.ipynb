# Stacked bar chart: Visualize the count of strokes by smoking status and gender.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('healthcare-dataset.csv')
df_stroke = df[df['stroke'] == 1]
stroke_counts = df_stroke.groupby(['smoking_status', 'gender']).size().unstack()
stroke_counts.plot(kind='bar', stacked=True)
plt.xlabel('Smoking Status')
plt.ylabel('Count of Strokes')
plt.title('Count of Strokes by Smoking Status and Gender')
plt.legend(title='Gender')
plt.show()
```

## Line plot: Analyze the trend of average glucose levels over age. 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from a CSV file into a DataFrame
df = pd.read_csv('dataset.csv')

# Group by age and calculate mean average glucose level
avg_glucose_by_age = df.groupby('age')['avg_glucose_level'].mean()

# Create a line plot of average glucose levels over age
plt.plot(avg_glucose_by_age.index, avg_glucose_by_age.values)
plt.xlabel('Age')
plt.ylabel('Average Glucose Level')
plt.title('Trend of Average Glucose Levels Over Age')
plt.show()
```

# Box plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset.csv')
plt.figure(figsize=(10, 6))
sns.boxplot(x='ever_married', y='avg_glucose_level', data=df)
plt.xlabel('Marital Status')
plt.ylabel('Average Glucose Level')
plt.title('Distribution of Average Glucose Levels by Marital Status')
plt.show()
```

# Group bar chat 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset.csv')
grouped_data = df.groupby('work_type').agg({'avg_glucose_level': 'mean', 'bmi': 'mean'})
```

```python
# Create a grouped bar chart for average glucose level and BMI
grouped_data.plot(kind='bar')
plt.xlabel('Work Type')
plt.ylabel('Value')
plt.title('Comparison of Average Glucose Level and BMI by Work Type')
plt.legend()
plt.show()
```
